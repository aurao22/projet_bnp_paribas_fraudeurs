```python
# 1. remplacement des valeurs 0 par np.nan
def fill_value_by_other(df, to_replace=0, replace_by = np.nan, verbose=0):
    """Replace all DF values = `to_replace` by `replace_by`

    Args:
        df (DataFrame): _description_
        to_replace (int or str), optional):Value to replace. Defaults to 0.
        replace_by (int or str), optional):Replacement value. Defaults to np.nan.
        verbose (int, optional): _description_. Defaults to 0.

    Returns:
        DataFrame: The updated DataFrame
    """
    short_name = "fill_zero_by_na"
    na_before = df.isna().sum().sum()
    m_before = df.memory_usage().sum() / 1e3
    for col in df.columns:
        df.loc[df[col]==to_replace, col] = replace_by
    if verbose>0:
        na_after = df.isna().sum().sum()
        info(short_name, f"{na_before} NA before \t--> {na_after} NA After --> \t {na_after-na_before} NA")
        m_after = df.memory_usage().sum() / 1e3
        info(short_name, f'{m_before:0.2f} bytes before \t--> {m_after:0.2f} bytes after --> \t {m_after-m_before} bytes')
    return df   
```
